# Polynomial Secret Sharing

```python
from sympy.abc import x, y, z
```

**Definition 2.1** 
A single variable polynomial with real coefficients is a function $f$ that takes a real number as input, produces a real number as ouput and has the form
$$
f(x) = a_0 + a_1x + a_2x^2 + ... + a_nx^n
$$
where $a_i$ are  real numbers. The $a_i$ are called *coefficients* of $f$. The *degree* of the polynomial is the integer $n$.

```python
# examples
expr1 = x**2 + 3*x + 2
expr1.as_poly()
```

```python
expr2 = (x + 2) * (3*x - 3)
expr2.as_poly()
```

```python
from IPython.display import display, Math
from sympy import latex
from sympy import degree

e1 = expr1.as_poly()
e2 = expr2.as_poly()

for e in (e1, e2):
    display(Math(rf"\text{{The expression }} {latex(e.as_expr())} "
                 rf"\text{{ has degree }} {e.degree()}"))
    display(Math(rf"\text{{Coefficients: }} {latex(e.all_coeffs())}"))
    print()

#print(degree(e1))
```

```python
from sympy import Poly

exp_poly = Poly(e1, x, domain='RR')
assert degree(exp_poly) == exp_poly.degree()
exp_poly
```

```python
from sympy import plot

# Figure 2.1

expr = x**5 - x - 1
f = Poly(expr, x, domain='ZZ')

# f(x) for -1.5 to 1.5
p = plot(f.as_expr(), (x, -1.5, 1.5), show=True)
```
